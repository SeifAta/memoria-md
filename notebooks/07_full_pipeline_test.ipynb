```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import json

from utils.llm import generate_structured
from utils.prompt_loader import load_prompt
from utils.examiner_schema import ExaminerReport
```

```python
with open(
    "../outputs/heart_failure_case.json",
    "r",
    encoding="utf-8"
) as f:

    patient_case = json.load(f)
```

```python
print(patient_case["metadata"]["title"])
```

```python
conversation = """
Student:
What brings you to the hospital today?

Patient:
Doctor, I just can't catch my breath anymore, and my legs are so swollen.

Student:
How long has this been happening?

Patient:
It has been getting worse over the past few weeks.

Student:
Do you sleep with extra pillows?

Patient:
Yes, I need three pillows because I become breathless when lying flat.

Student:
Do you have any medical conditions?

Patient:
I have high blood pressure and diabetes.
"""
```

```python
prompt = load_prompt(
    "examiner_agent.md",
    PATIENT_CASE=json.dumps(
        patient_case,
        indent=2
    ),
    CONVERSATION=conversation
)
```

```python
report = generate_structured(
    prompt,
    ExaminerReport
)
```

```python
print(
    "Overall Score:",
    report.overall_score
)


print("\nMissed Points:")
for point in report.history_assessment.missed_points:
    print("-", point)


print("\nRecommendations:")
for item in report.key_recommendations:
    print("-", item)
```
