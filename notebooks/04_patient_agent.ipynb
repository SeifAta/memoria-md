```python
import json

from utils.llm import generate
from utils.prompt_loader import load_prompt
```

```python
with open(
    "../outputs/heart_failure_case.json",
    "r",
    encoding="utf-8"
) as f:

    patient_case = json.load(f)
```

```python
conversation = []


def ask_patient(question):

    conversation.append(
        f"Student: {question}"
    )

    history = "\n".join(conversation)

    prompt = load_prompt(
        "patient_agent.md",
        PATIENT_CASE=json.dumps(
            patient_case,
            indent=2
        ),
        QUESTION=history
    )

    response = generate(prompt)

    conversation.append(
        f"Patient: {response}"
    )

    return response
```

```python
print(
    ask_patient(
        "What brings you to the hospital today?"
    )
)
```

```python
print(
    ask_patient(
        "How long has this been happening?"
    )
)
```

```python
print(
    ask_patient(
        "Has it become worse?"
    )
)
```
