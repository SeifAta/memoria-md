```python
from utils.pdf import extract_pdf_text
from utils.prompt_loader import load_prompt
from utils.llm import generate_structured
from utils.schemas import TeachingBlueprint

from pathlib import Path
import json
```

```python
lecture = extract_pdf_text(
    Path("../sample_data/heart_failure.pdf")
)

print(lecture[:1000])
```

```python
prompt = load_prompt(
    "teaching_blueprint.md",
    LECTURE=lecture
)
```

```python
blueprint = generate_structured(
    prompt,
    TeachingBlueprint
)
```

```python
output_path = Path("../outputs/heart_failure_blueprint.json")

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(
        blueprint.model_dump(),
        f,
        indent=2,
        ensure_ascii=False
    )

print("Saved!")
```
