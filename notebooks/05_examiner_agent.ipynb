```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import json

from utils.llm import generate_structured
from utils.prompt_loader import load_prompt
from utils.examiner_schema import ExaminerReport
```

```python
with open(
    "../outputs/heart_failure_case.json",
    "r",
    encoding="utf-8"
) as f:

    patient_case = json.load(f)
```

```python
conversation = """
Student:
What brings you to the hospital today?

Patient:
Doctor, I just can't catch my breath anymore, and my legs are so swollen.

Student:
How long has this been happening?

Patient:
It has been getting worse over the past few weeks.

Student:
Do you sleep with extra pillows?

Patient:
Yes, I need three pillows because I become breathless when lying flat.
"""
```

```python
prompt = load_prompt(
    "examiner_agent.md",
    PATIENT_CASE=json.dumps(
        patient_case,
        indent=2
    ),
    CONVERSATION=conversation
)


report = generate_structured(
    prompt,
    ExaminerReport
)


print(report)
```

```python
import json
from pathlib import Path

output_path = Path("../outputs/examiner_report.json")

with open(
    output_path,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        report.model_dump(),
        f,
        indent=2
    )

print("Saved:", output_path)
```
